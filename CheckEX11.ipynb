```python
# pandas関連ライブラリの導入
import pandas as pd
# データフレーム表示用関数
from IPython.display import display

# データフレームでの表示精度
pd.options.display.float_format = '{:.4f}'.format
# データフレームですべての項目を表示
pd.set_option("display.max_columns",None)
```

```python
# チェック演習11-1：
# (1) 次のリストは、ある年の「地価公示　区市町村別用途別地点数」で、縦の列項目は「区」,「住宅地」,「商業地」,「全用途」です。
# データフレームdf11-1を作り、表示する。
# [千代田区, 7, 52, 59], [中央区,   8, 55, 63], [港  区,  30, 51, 81], [新宿区,  27, 49, 76], [渋谷区,  26, 33, 59]

# リストの定義
data = [
    ['千代田区', 7, 52, 59],
    ['中央区',  8, 55, 63],
    ['港区', 30, 51, 81],
    ['新宿区', 27, 49,76],
    ['渋谷区', 26, 33,59]]

# 結果確認
print(data)

# データフレームの定義
df11A = pd.DataFrame( data, columns= ['区', '住宅地', '商業地', '全用途'] )

# 型表示
print(type(df11A))

# display関数による整形表示
display(df11A)
```

```python
# チェック演習11-1：
# (2) 作成したデータフレームの内容を、メソッドcolumns、index、valuesで再表示
# 列インデックスと行インデックス
print('列インデックス', list(df11A.columns));print()
print('行インデックス', list(df11A.index));print()
print('データフレームの値')
print(df11A.values)
print()
# (3) 作成したデータフレームの第1行を抽出して表示
print('第1行', df11A.values[1:2, ])
```

```python
# チェック演習11-2：
#チェック演習11-1のデータフレームdf11Aを使って(1)、(2)を行う。
# (1) 第0列を抽出して、Series変数srow0を作る。

# データフレームの特定列からSeries変数を抽出
scol0 = df11A['区']
print(type(scol0))
print(scol0)

# (2) 第3列を抽出して、Series変数scol3を作る。
# データフレームの特定行からSeries変数を抽出
srow3 = df11A.loc[3]
print(type(srow3))
print(srow3)
```

```python
# チェック演習11-3：
#チェック演習11-1のデータフレームdf11Aを使って(1)、(2)、(3)を行う。
# (1) データ型を確認する。
print(df11A.dtypes)

# (2) 列の一部（第0，2列）を抽出したデータフレームdf11Bを作り、確認する。
# 抽出したい項目名のリストを準備
column1 = ['区', '商業地']
#  変数名[リスト変数]の形で列を抽出
df11B = df11A[column1]
# 結果確認
display(df11B)

# (3) 行の一部（第1，2行）を抽出したデータフレームdf11Cを作り、確認する。
df11C = df11A.loc[1:2]
display(df11C)
```

```python
# チェック演習11-4：
# チェック演習11-1のデータフレームdf11Aを使って(1)、(2)を行う。
# (1) 最初から3行を表示する。
display(df11A.head(3))
# (2) 最後から3行を表示する。
display(df11A.tail(3))
```

```python
# チェック演習11-5：
# チェック演習11-1のデータフレームdf11Aを使って、「商業地」が50以下の行を抽出し、データフレームdf11Dとし、表示する。
# 「平均が70以上]の判断(短縮して1行で行う)
df11D = df11A[df11A['商業地'] <= 50]

# 表示
display(df11D)
```

```python
# チェック演習11-6：
# チェック演習11-1のデータフレームdf11Aを使って
# (1) 第1列を削除したデータフレームをdf11Eとし、表示する。
df11E = df11A.drop('住宅地', axis=1)
# 結果を確認
display(df11E)

# (2) 第1行を削除したデータフレームをdf11Fとし、表示する。
df11F = df11A.drop(1, axis=0)
# 結果を確認
display(df11F)
```

```python
# チェック演習11-7：
# チェック演習11-1のデータフレームdf11Aを使って、'住宅地', '商業地', '全用途'の'合計点数'を求め、
# 第1列に挿入したデータフレームdf11Gを作り、表示する。
# 合計点の計算
s1 = df11A['住宅地'] + df11A['商業地'] + df11A[ '全用途']

# 結果確認
print(s1)

# データコピー
df11G = df11A.copy()

# 列追加(insert関数で任意の位置の挿入)
df11G.insert(1, '合計点数', s1)

# 結果確認
display(df11G)
```

```python
import numpy as np
# チェック演習11-8：
# チェック演習11-1のデータフレームdf11Aを使って、次のデータ列を追加したデータフレームdf11Hを作り、表示する。
# 新しいデータ
s2 = pd.Series(np.array([3.0, 4.0, 5.0, 6.0, 7.0]), name='不明率%')

# 既存データフレームに新しい列を追加
df11H = pd.concat([df11A, s2], axis=1)
# 結果確認
display(df11H)
```

```python
# 追加する学生のデータ
# 追加する学生のデータ
new_row = {'区': '文京区', '住宅地': 30, '商業地': 70, '全用途': 50}
display(new_row)
# 既存データフレームに追加
df11I = pd.concat([df11A, pd.DataFrame([new_row])], ignore_index=True)

# 結果確認
display(df11I)
```
