本章でのチェック演習は、自分で用意したCSVファイル使って練習を行うことが目的ですが、模範解答を示すため、「dataCE12.csv」を用意します。このデータは、ある年度の各都市の幼稚園、小学校、中学校の情報です。

```python
# *** 事前設定 ***
# 関連ライブラリの導入
import numpy as np
import pandas as pd
# データフレーム表示用関数
from IPython.display import display

# numpy表示形式を設定
np.set_printoptions(suppress=True, precision=1, floatmode='fixed')

# pandasデータフレームでの表示精度
pd.options.display.float_format = '{:.1f}'.format
# pandasデータフレームですべての項目を表示
pd.set_option("display.max_columns",None)
```

```python
# チェック演習12-1：
# Google Colaboratoryに、欠損なしデータ「dataCE12.csv」をアップロードし、
# 最初と終わりの5行を表示して、内容を確認する。

# CSVファイルの読み込み(欠損データなし)
df1 = pd.read_csv(
    '/content/drive/My Drive/Colab Notebooks/Python_Txt_toshimitsu/Ver1.0_checkEx/dataCE12.csv', header = 0)
display(df1.head())
display(df1.tail())
```

```python
# チェック演習12-2：
# Google Colaboratoryに、欠損のあるデータ「dataCE12_D.csv」をアップロードし、
# 最初と終わりの5行を表示して、内容を確認する。

df1D = pd.read_csv(
    '/content/drive/My Drive/Colab Notebooks/Python_Txt_toshimitsu/Ver1.0_checkEx/dataCE12_D.csv', header = 0
    # , encoding='shift_jis'
    )
display(df1D.head())
display(df1D.tail())
```

```python
# チェック演習12-3：
#「dataCE12.csv」のindexを「地域コード」にした新たなデータフレームdf2を作る。
df2D = df1D.set_index('地域コード')
display(df2D.head())
```

```python
# チェック演習12-4：
# チェック演習12-3で作ったdf2DをCSVファイル「outdataCE12-4,csv」で書き出し、
# Google Colaboratory内にあることを確認する。
df2D.to_csv('outdataCE12-4.csv', index = False)
!ls -l -F
```

```python
# チェック演習12-5：
# チェック演習12-3で作ったdf2DをExcelファイル「outdataCE12-4,xlsx」で書き出し、
# Google Colaboratory内にあることを確認する。
df2D.to_excel('outdataCE12-4.xlsx', index = False)
!ls -l -F
```

```python
# チェック演習12-6：
# チェック演習12-3で作ったdf2Dの列項目「小学校教員数」,「中学校教員数」の列データを削除し、
# 列項目名を自分で変更したdf4Dを作り、確認する。

# 加工用の別データフレームdf3Dの用意
df3D = df2D.copy()
# 項目「小学校教員数」と「中学校教員数」を削除し、結果をdf4Dに代入
df4D = df3D.drop(['小学校教員数', '中学校教員数'], axis=1)
# 結果を確認
display(df4D)

# 差し替え用項目名リスト
# df4Dは8列なので、colsリストも8要素
cols = [
    'pref.', 'city', '幼稚園数','幼稚園在園者数','小学校数','小学校児童数','中学校数', '中学校生徒数'
]
# 列名の差し替え
df4D.columns = cols
# 結果確認
display(df4D.head())
```

```python
# チェック演習12-7：
# チェック演習12-6で作ったdf4Dの欠損値を確認する。
df4D.isnull().sum()
```

```python
# チェック演習12-8：
# チェック演習12-6で作ったdf4Dの欠損値を0の置き換えたデータフレームdf5Dを作る。
# fillnaの引数(0)は、置き換える値
df5D = df4D.fillna(0)
display(df5D)
```

```python
# チェック演習12-9：
# チェック演習12-8で作ったデータフレームdf5Dの'幼稚園数','幼稚園在園者数'
# をコード12.4.6の方法（columns, removeを使う）で削除する。
df6D = df5D.copy()
# 列名のリストを作成
columns_new = list(df6D.columns)
# 各項目をリストから個別に削除
columns_new.remove('幼稚園数'); columns_new.remove('幼稚園在園者数')
# 削除した項目を含まない列リストで新しいデータフレームを作成
print(columns_new)
df6D = df6D[columns_new]
display(df6D.head())
```

```python
# チェック演習12-10：
# チェック演習12-9で作ったデータフレームdf6Dの列項目を、「中学校数」、「中学校生徒数」、
# 「小学校数」、「小学校児童数」の順番にしたデータフレームdf7Dを作る。
df7D = df6D.copy()
columns1 = list(df7D.columns)
print(columns1)
columns2 = columns1[0:2] + columns1[4:] + columns1[2:4]
df7D = df7D[columns2]
df7D.head()
```

```python
# チェック演習12-11：
# チェック演習12-8で作ったデータフレームdf5Dのデータ型の再確認し、
# 「幼稚園在園者数」をint型に変換したデータフレームdf8Dを作る。
display(df5D.dtypes)

df8D = df5D.copy()
df8D['幼稚園在園者数'] = df8D['幼稚園在園者数'].astype('int')
display(df8D.dtypes)
display(df8D.head())
```

```python
# チェック演習12-12：
# チェック演習12-11で作ったデータフレームdf8Dの数値データの統計量を調べる。

# 引数なしのdescribeメソッド呼び出し
df8D.describe()
```

```python
# チェック演習12-13：
# チェック演習12-11で作ったデータフレームdf8Dで、都道府県ごとの都市数を調べる。

# 値の出現章数カウント
df8D['pref.'].value_counts()
```
