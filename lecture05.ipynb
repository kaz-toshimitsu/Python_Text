5.3　定義関数の実装

```python
# コード5.3.1　引数１，戻り値１の関数の実装
# *** 定義関数「func1」***
def func1(x):
    y = 1 + 2 * x + x**2
    return y

```

```python
# コード5.3.2　引数2，戻り値2の関数の実装
# *** 定義関数「func22」***
def func22(x1, x2):
    y1 = 1 + 2 * x1 + x1**2
    y2 = x1 - x2
    return y1, y2
```

```python
# main program(呼び出し側のプログラム)
a = 5
b = 10
# 定義関数「func22」の呼び出し
print('a = ',a, '  b =', b)
print(func22(a, b))
```

```python
# コード5.3.3　returnなしの関数の実装
# *** 定義関数「func3」***
def func3(x):
    print(type(x))
    print(x)
```

```python
# main program(呼び出し側のプログラム)
x = 12345

# 定義関数「func3」の呼び出し
func3(x)
```

```python
# コード5.3.4　引数でない変数の定義関数と呼び出しプログラムの独立性
# *** 定義関数「func4」***
def func4(x):
    a = 10
    print('func4 の a =',a)
```

```python
# main program(呼び出し側のプログラム)
x = 12345
a = 1
# 定義関数「func4」の呼び出し
func4(x)

# main progrmuのa は定義関数内のaとは関係しない
print('main の a =',a)
```
