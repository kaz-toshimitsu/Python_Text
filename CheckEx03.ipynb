第3章　チェック演習

```python
# チェック演習3-1：ifを使って、変数xを100として、「xは100以上」とxの値を表示する。
x = 100
if x >= 100:
    print('xは100以上'); print(x)
```

```python
# チェック演習3-2：ifを使って、変数xを50として、xが100未満で「xは100未満」と100以上で「xは100以上」を場合分けして表示し、さらにどちらの場合でもxの値を表示する。
x = 50
if x < 100:
    print('xは100未満')
else:
    print('xは100以上')

print(x)
```

```python
# チェック演習3-3：a=10,b=20として、a,bがどちらも5以上であれば、「a,bは両方5以上」とa,bの値を表示する。そうでなければなにも表示しない。
a = 10; b = 20
if (a >= 5 and b >= 5):
    print('a,bは両方5以上')
    print(a, b)
```

```python
# チェック演習3-4：x=10として、3で割った余りが0, 1, 2に場合分けして「余りは０」、「余りは１」、「余りは２」の結果を表示する。
x = 10
y = x % 3

if y == 0:
    print('余りは０')
elif y == 1:
    print('余りは１')
else:
    print('余りは２')
```
