10.3　散布図（scatter関数）

```python
# コード10.3.1　描画データ「アイリスデータセット」の準備
# グラフ描画用のデータを、ライブラリを用いて取得する
from sklearn.datasets import load_iris
iris = load_iris()
x, y = iris.data, iris.target
columns = iris.feature_names

# x, y, columnsの型を確認
print(type(x), type(y), type(columns))
```

```python
# コード10.3.2　描画データ「アイリスデータセット」の確認
# 読み込んだデータの確認
# columnsの内容
print('column:',columns)
# xのshape確認と先頭5行
print('xのデータ型: ', x.shape)
print('x', x[:5])

# yのhape確認と先頭5行
print('yのデータ型: ', y.shape)
print('y', y[:5])
```

```python
# コード10.3.3　簡単な散布図
# 簡単な散布図

# scatter関数呼び出し
plt.scatter(x[:,0], x[:,1])

# 描画
plt.show()
```

```python
# コード10.3.4　散布図を見やすくする
# scatter関数呼び出し（yの値で色を変える）
plt.scatter(x[:,0], x[:,1], c=y, cmap='rainbow')

# 方眼表示
plt.grid()

# 軸ラベル表示
plt.xlabel(columns[0])
plt.ylabel(columns[1])

# タイトル表示
plt.title('アイリスデータセットによる散布図')

# 描画
plt.show()
```

10.4　x-yプロット

```python
# コード10.4.1　x-yプロット描画データ
# ｘの配列
x = np.linspace(0, 1, 11)
print(x)

# y = exp(x)の計算
y = np.exp(x)
print(y)
```

```python
# コード10.4.2　x-y プロット
plt.plot(x, y)
# 描画
plt.show()
```

```python
# コード10.4.3　x-yプロット（2関数の重ね書き）
# ｘの配列
x = np.linspace(0, 1, 11)

# y = exp(x)の計算
y1 = np.exp(x) # 指数関数
y2 = 1 + x**2      # 2次関数
```

```python
# plot関数を2回続けて呼び出すと重ね描きになる
# label引数は、凡例表示をする場合に必要
plt.plot(x, y1, label='指数関数[exp(x)]')
plt.plot(x, y2, label='2次関数[1+x**2]')

# 凡例表示
plt.legend()

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('2つの関数の重ね描き')
plt.show()
```

```python
# コード10.4.4　複数グラフの描画のデータ準備
# データの準備（描画データ「アイリスデータセット」ですでに説明済み）
from sklearn.datasets import load_iris
iris = load_iris()
x, y = iris.data, iris.target
columns = iris.feature_names
print(x[:5])
```

```python
# コード10.4.5　複数グラフの描画（subplot関数）
# (1)サイズ指定
plt.figure(figsize=(8, 4))

# (2),(3)グラフ数回ループを回す
for i in range(1, 3):

    # i 番目のax変数取得
    ax = plt.subplot(1, 2, i)

    # 散布図表示
    ax.scatter(x[:,0], x[:,i], c=y, cmap='rainbow')

    # タイトル表示
    ax.set_title(columns[0] + ' vs ' + columns[i])

    ax.grid()
# (4)隣接オブジェクトと重ならないようにする
plt.tight_layout()

# 表示
plt.show()
```
