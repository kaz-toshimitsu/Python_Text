pandas・numpyの初期設定

```python
# 必要ライブラリのimport
import numpy as np

# コード11.2.1 pandas関連ライブラリ
import pandas as pd

# データフレーム表示用関数
from IPython.display import display
```

```python
# numpy表示形式の設定
np.set_printoptions(suppress=True, precision=4, floatmode='fixed')
```

```python
# コード11.2.2 pandas関連パラメータ設定
# データフレームでの表示精度
pd.options.display.float_format = '{:.4f}'.format

# データフレームですべての項目を表示
pd.set_option("display.max_columns",None)
```

問題11ー1: 指定の表のデータフレームを定義しなさい。

```python
# 2重リスト変数の定義
data_A = [
    ['24S1001', 1980, 'male', 100.0],
    ['24S1002', 2000, 'female', 900.0],
    ['24S1003', 1998, 'male', 1200.0],
    ['24S1004', 2023, 'male', 600.0]
]

# 結果確認
print(data_A)
# データフレームの定義
df_A = pd.DataFrame( data_A,columns=['ID', ' access ', ' gender ', ' budget '] )

# 型表示
print(type(df_A))
print()

# display関数による整形表示
display(df_A)
```

問題11-2：
問題11-1で作成した、データフレームの列インデックス、行インデックス、データフレームの値を表示しなさい。


```python
# データフレームの各部品表示

# 列インデックスと行インデックス
print('列インデックス', list(df_A.columns))
print('行インデックス', list(df_A.index))
print()

# データフレームの値
print('データフレームの値')
print(df_A.values)
```

問題11-3: 問題11-1で作成した、データフレームの最初の１行と最後の1行を表示しなさい。


```python
# head(n)メソッドで先頭n行の抽出（指定しない場合は、デフォルト値=5）
display(df_A.head(1))

# tail(n)メソッドで最後のn行の抽出（指定しない場合は、デフォルト値=5）
display(df_A.tail(1))
```

問題11-4: 問題11-1で作成した、データフレームの第1列(access)と第2列(gender)の間に下記のデータを差し込みなさい。


```python
# 2重リスト変数の定義
data_B = [
    [2024],
    [2023],
    [2024],
    [2024]
]
# 結果確認
print(data_B)
# データフレームの定義
df_B = pd.DataFrame( data_B,columns=['access1 '] )

# 型表示
print(type(df_B))
print()
# display関数による整形表示
display(df_B)

# データコピー
df_G = df_A.copy()

# 列追加(insert関数で任意の位置の挿入)
df_G.insert(2, 'access1', df_B)

# 結果確認
display(df_G)
```
