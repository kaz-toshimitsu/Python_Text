本編データフレーム「df2」の準備

```python
# コード13.2.1　CSVファイルの読み込み(欠損データなし)

df1 = pd.read_csv(
    '/content/drive/My Drive/Colab Notebooks/Python_Txt_toshimitsu/data12-2_modify.csv',
    header = 1, index_col = 'データID'
)

# コード13.2.2　データフレームの列の項目名変更
# 差し替え用項目名リスト
# 列の項目名定義
cols = [
    '性別', '地域コード',  '都道府県', '人口（10歳以上）',
    '学習・自己啓発',  '学習・自己啓発_方法', '英語', '英語以外の外国語',
    '好きな勉強法',  '商業実務(総数)'
]
# 加工用の別データフレームの用意
df2 = df1.copy()
# 列名の差し替え
df2.columns = cols
# 結果確認
display(df2.head())
```

問題13-1:　本編で作成したデータフレーム「df2」の項目「学習・自己啓発_方法」で、「groupbyメソッド」を使ってグループ分けし、「学習・自己啓発」の項目で集計結果を表示してください。



```python
# グループによる集計結果
df2_gr2 = df2.groupby(['学習・自己啓発_方法'])['学習・自己啓発'].sum()

# 結果確認
display(df2_gr2)
```

問題13-2：　「学習・自己啓発_方法」と「好きな勉強法」を軸とした、「商業実務(総数)」結果のクロス集計を行ってください。


```python
# 「学習・自己啓発_方法」と「好きな勉強法」を軸とした、「商業実務(総数)」結果のクロス集計
df_pivot = df2.pivot_table(
    index='学習・自己啓発_方法',
    columns='好きな勉強法',
    values='商業実務(総数)',
    aggfunc='sum'
)

# 結果確認
display(df_pivot)
```
