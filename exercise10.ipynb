問題１０ー１

```python
# sigmoid関数の定義
def sigmoid(x):
  return 1 / (1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
# plt.ylim(-0.1, 1.1)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('シグモイド関数')

plt.show()
```

問題10-2:
コード10.4.5では「複数グラフの描画」で、「がく片（sepal） の長さ - がく片（sepal） の幅」と「がく片（sepal）の長さ - 花弁（petal）の長さ」の二つのグラフを同時に表示しました。ここでは、「がく片（sepal） の長さ - がく片（sepal） の幅」と「がく片（sepal）の長さ - 花弁（petal）の幅」の二つのグラフを同時に表示しなさい。


```python
# 複数グラフの描画（subplot関数）
# データの準備（描画データ「アイリスデータセット」ですでに説明済み）
from sklearn.datasets import load_iris
iris = load_iris()
x, y = iris.data, iris.target
columns = iris.feature_names
print(x[:5])
```

```python
# サイズ指定
plt.figure(figsize=(8, 4))

# 2つのグラフの描画

# ax変数取得
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

# 散布図表示
ax1.scatter(x[:,0], x[:,1], c=y, cmap='rainbow')
ax2.scatter(x[:,0], x[:,3], c=y, cmap='rainbow')

# タイトル表示
ax1.set_title(columns[0] + ' vs ' + columns[1])
ax2.set_title(columns[0] + ' vs ' + columns[3])

ax1.grid()
ax2.grid()
# 隣接オブジェクトとぶつからないようにする
plt.tight_layout()

# 表示
plt.show()
```
