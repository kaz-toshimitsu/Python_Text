8.2　ライブラリの宣言（import文）

```python
# コード8.2.1　標準ライブラリmathの宣言と使用例(math.sqrt)
import math

# mathライブラリのsqrt関数は「math.sqrt」の形で使用する
x = math.sqrt(4)
print(x)
```

```python
# コード8.2.2　標準ライブラリmathの宣言と使用例(sqrt)
# ライブラリ中の特定の関数のみ利用する場合は下記の方法でも可
from math import sqrt

# 「sqrt」だけで関数を呼び出せる
x = sqrt(4)
print(x)
```

```python
# コード8.2.3　外部ライブラリnumpyを利用する宣言と使用例
# 「as np」は別名での定義/ numpyの関数は「np.関数名」の形で使用する
import numpy as np
x = [0, 1, 2, 3, 4, 5]

# numpyのsqrt関数の呼び出し
y = np.sqrt(x)
# 結果確認 (分割した各値の平方根を一括で計算する)
print(y)
```
