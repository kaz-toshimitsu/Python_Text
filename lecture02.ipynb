2.5　データの型と四則演算

(1)　データの型とtype関数

```python
# コード2.5.1　整数型の確認
# 変数aに 整数型の値321を代入する
a = 321

# 値を確認する
print(a)

# 型を確認する
print(type(a))
```

```python
# コード2.5.2　浮動小数点数型の確認
# 変数bに 浮動小数点数型の値3.14を代入する
b = 3.14

# 値を確認する
print(b)

# 型を確認する
print(type(b))
```

```python
# コード2.5.3　文字列型の確認
# 変数cに 文字列型の値‘Hello’を代入する
c = 'Hello'

# 値を確認する
print(c)

# 型を確認する
print(type(c))
```

```python
# コード2.5.4　ブール型の確認
# 変数dに ブール型の値‘True’を代入する
d = True

# 値を確認する
print(d)

# 型を確認する
print(type(d))
```

2.2.2　四則演算

```python
# コード2.5.5　整数型と浮動小数点数型の足し算
# 変数i1に 整数型の値100を代入する
# 変数f1に 整数型の値50.55を代入する
i1 = 100
f1 = 50.55

f2 = i1 + f1

print(f2)
print(type(f2))
```

```python
# コード2.5.6　文字列型の足し算
# 変数s1に文字列型の'I&'を代入する
# 変数s2に文字列型の'You'を代入する
s1 = 'I&'
s2 = 'You'

s3 = s1 + s2

print(s3)
print(type(s3))
```

```python
# コード2.5.7 文字列型と整数型の足し算はできない!
# 変数s1に文字列型の'I&'を代入する
# 変数i1に 整数型の値100を代入する
s1 = 'I&'
i1 = 100

s4 = s1 + i1

print(s4)
print(type(s4))
```

```python
# コード2.5.8　整数型と浮動小数点数型のかけ算
# 変数i1に 整数型の値100を代入する
# 変数f1に 整数型の値50.55を代入する
i1 = 100
f1 = 50.55

f3 = i1 * f1

print(f3)
print(type(f3))
```

```python
# コード2.5.9　整数型の割り算
# 変数i1に 整数型の値100を代入する
# 変数i2に 整数型の値24を代入する
i1 = 100
i2 = 24

f4 = i1 / i2

print(f4)
print(type(f4))
```

```python
# コード2.5.10　整数割り算と余り
# 変数i1に 整数型の値10を代入する
# 変数i2に 整数型の値3を代入する
i1 = 10
i2 = 3
# 整数の割り算で、結果を整数で打ち切る
i3 = i1 // i2
# 整数同士の割り算で余りを求める
i4 = i1 % i2

print(i3)
print(type(i3))

print(i4)
print(type(i4))
```

```python
# コード2.5.11　整数型のべき乗
# 変数i1に 整数型の値10を代入する
# 変数i2に 整数型の値3を代入する
i1 = 10
i2 = 3
# 累乗計算
i5 = i1 ** i2

print(i5)
print(type(i5))
```

```python
# コード2.5.12　浮動小数点数型のべき乗
# 変数f1に 浮動小数点数型の値10.0を代入する
# 変数f2に 浮動小数点数型の値3.0を代入する
f1 = 10.0
f2 = 3.0
# 累乗計算
f5 = f1 ** f2

print(f5)
print(type(f5))
```

```python
# コード2.6.1　累算代入演算子
# 変数 z に 10を入れる
z = 10

# 変数zに z + 5 とした値を入れ直し、参照する
z += 5    # z = z + 5
print(z)
```
