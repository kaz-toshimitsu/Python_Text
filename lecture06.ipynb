6.4　リストの要素の参照（スライス）

```python
# コード6.4.1 リスト型の要素の参照
#
L = [2, 4, 6, 8, 10]
print(' L[0] =', L[0])
print(' L[1] =', L[1])
print(' L[2] =', L[2])
print(' L[3] =', L[3])
print(' L[4] =', L[4])

# 最後の要素のインデックスは-1でもよい（ー1は後ろから1番目の意味）
print(' L[-1] =', L[-1])
```

```python
# コード6.4.2 スライスによる要素範囲の指定
# リストの定義
L = [2, 4, 6, 8, 10]
# L の全要素
print('L の全要素', L[:])   # print（L)でもよい

# インデックス1以上5未満
print(L[1:5])

# 最初のインデックスを省略すると[0]指定したのと同じ
print(L[:3])

# 2番目のインデックスを省略すると「最後の要素まで」の意味になる
print(L[3:])

# 2番目のインデックスを[-1]にすると「最後の要素を除いた残り全ての要素」の意味になる
print(L[:-1])
```

6.5　演算や操作

```python
# コード6.5.1 アンパック
# 文字型リストの定義
str4 = ['阿部', '井上', '植木', '江藤']

# リストのアンパック
s0, s1, s2, s3 = str4

# アンパック後の各変数の確認
s0, s1, s2, s3
```

```python
# コード6.5.2 「＋」加法演算子
# 文字型リストの定義
L = [2, 4, 6, 8, 10]
str4 = ['阿部', '井上', '植木', '江藤']
# 全要素の表示
print('L    の全要素', L)
print('str4 の全要素', str4)

# 加法演算
LS = L + str4
# 確認
LS
```

```python
# コード6.5.3 len関数
# 文字型リストの定義
str4 = ['阿部', '井上', '植木', '江藤']

# 要素数を数える　len関数
n = len(str4)
# 確認
n
```

```python
# コード6.5.4 appendメソッド
# 文字型リストの定義
# 初期値として、空のリストを用意
str_a = []

# appendメソッドで、'大川', '加藤'を追加
str_a.append('大川')
str_a.append('加藤')

# 確認
str_a
```

```python
# コード6.5.5 in関数
# 文字型リストの定義
str4 = ['阿部', '井上', '植木', '江藤']
# リストにあるか調べる指定要素
s1 = '井上'
s2 = '斉藤'

# in関数でリスト「str4」に指定した要素「s1, s2」があるか調べる
a = s1 in str4
b = s2 in str4

# 確認
print(str4)
print('"井上"はリストにある：', a)
print('"斉藤"はリストにある：', b)
```

```python
# コード6.5.6 if文の条件としてin関数を利用する例
# 文字型リストの定義
str4 = ['阿部', '井上', '植木', '江藤']
# リストにあるか調べる指定要素
s1 = '植木'

# if文の条件にin関数を使用
if s1 in str4:
    print(str4, 'の中に', s1, 'がある')
```

6.6　ループ処理（in関数, range関数）

```python
# コード6.6.1 ループ処理にin関数を利用する例
# 文字型リストの定義
str4 = ['阿部', '井上', '植木', '江藤']

# for文にin関数を使用
for s in str4:
    print(s)
```

```python
# コード6.6.2 ループ処理にrange関数を利用する例
# 文字型リストの定義
str4 = ['阿部', '井上', '植木', '江藤']
n = len(str4)

# for文にrange関数を使用
for i in range(n):
    print(i, str4[i])
```

6.7　ループ処理（enumerate関数, 内包表記）

```python
# コード6.7.1 enumerate関数
# リスト型変数の定義
n1   = [11, 22, 33, 44, 55, 66]
str4 = ['阿部', '井上', '植木', '江藤']
s1   = '阿部真一'

# enumerate関数処理の確認
print(list(enumerate(n1)))
print(list(enumerate(str4)))
print(list(enumerate(s1)))
```

```python
# コード6.7.2 enumerate関数を利用したループ処理
# リスト型変数の定義
str4 = ['阿部', '井上', '植木', '江藤']

# iはインデックス、sはリストの各要素、
for i, s in enumerate(str4):
    print(i, '番目の要素 =', s)
```

```python
# コード6.7.3 内包表記を利用したループ処理
# 処理するリスト型変数の定義
# 数値要素のリスト
list_base   = [10, 20, 30, 40, 50, 60]

# 内包表記を用いて新しいリストを作るループ処理
# xはリストの各要素で、x**2 (xの2乗）の数式を各要素数値に適用
list_new = [x**2 for x in list_base]
print(list_new)
print(type(list_new))
```

```python
# コード6.7.4 内包表記を利用したループ処理
# 処理するリスト型変数の定義
# 文字列要素のリスト
str4 = ['阿部', '井上', '植木', '江藤']

# 内包表記を用いて新しいリストを作るループ処理
# xはリストの各要素で、x[0]は各要素中の0番目文字を指定
str_0 = [x[0] for x in str4]
print(str_0)
print(type(str_0))
```
