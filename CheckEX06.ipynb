第６章　チェック演習

```python
# チェック演習6-1：要素が文字型で8個のあるリストを定義して表示し、型と確認する。
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
print(L)
print(type(L))
```

```python
# チェック演習6-2：チェック演習6-1のリストLのインデックス2と一番後ろの要素表示する。
print('L[ 2] = ', L[2])
print('L[-1] = ', L[-1])
```

```python
# チェック演習6-3：チェック演習6-1のリストLのインデックス2から4をスライスして保存し、表示する。
L24 = L[2:5]
print(L24)
```

```python
# チェック演習6-4：チェック演習6-3のリストをアンパックして表示する。
L1, L2, L3 = L24
print(L1)
print(L2)
print(L3)
```

```python
# チェック演習6-5：チェック演習6-1のリストと6-3のリストを連結して表示する。
L65 = L + L24
print(L65)
```

```python
# チェック演習6-6：チェック演習6-5のリストの要素数を求め、表示する。
n = len(L65)
print(n)
```

```python
# チェック演習6-7：チェック演習6-5のリストに、'apple'と'orange'の要素数追加し、表示する。
L67 = []
L67 = L65 + ['apple', 'orange']
print(L67)
```

```python
# チェック演習6-8：チェック演習6-7のリストに、’x’，'apple'があるか調べ、表示する。
a = 'x' in L67
b = 'apple' in L67
print('"x" はリストにある = ', a)
print('"apple" はリストにある = ', b)
```

```python
# チェック演習6-9：チェック演習6-7のリストに、自分で指定する文字列があるか調べ、ある場合とない場合で場合分けし、表示する。
# リストにあるか調べる指定要素
s= 'cherry'

# if文の条件にin関数を使用
if s in L67:
    print(L67, 'の中に', s, 'がある')
else:
    print(L67, 'の中に', s, 'がない')
```

```python
# チェック演習6-10：コード6.6.1のリストstr4 = str4 = ['阿部', '井上', '植木', '江藤']の全要素の後ろに”（アパートLの住人）”を付けて、表示する。
str4 = ['阿部', '井上', '植木', '江藤']

for s in str4:
    new_s= s + "_（アパートLの住人）"
    print(new_s)
```

```python
# チェック演習6-11：コード6.6.2のリストstr3 = ['apple', 'orange', 'cherry']の全要素の前にインデックス番号を付けて、表示する。
# 備考：enumerate関数を使えば、簡単にできますが、プログラム練習のためenumerate関数を使わずにやってみる。
str3 = ['apple', 'orange', 'cherry']
s611 =[]
n = len(str3)

for i in range(n):
    s611 = str(i) + "_" + str3[i]
    print(s611)
```

```python
# チェック演習6-12：リストL = [10, 20, 30, 40, 50]の各全要素と「インデックス番号＋１」の掛け算して、表示する。
L = [10, 20, 30, 40, 50]

for i, s in enumerate(L):
    L612 = (i+1) * s
    print(i+1,' * ', s, ' = ', L612)
```

```python
# チェック演習6-13：チェック演習6-12のリストLの要素をfloat型にする。
L = [10, 20, 30, 40, 50]

# 内包表記を用いて新しいリストを作るループ処理
list_new = [ float(x) for x in L]
print(list_new)
```
