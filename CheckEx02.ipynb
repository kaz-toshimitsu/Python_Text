```python
# チェック演習2-1：変数a,bに、それぞれ10, 20 を入れて、10, 20 を表示しなさい。
a = 10
b = 20
print(a,b)
```

```python
# チェック演習2-2：【足し算】変数xに、100を入れ、それに20 を足した計算をしてyに入れ表示する
x = 100
y = x + 20
print(y)
```

```python
# チェック演習2-3：変数xに、100を入れ、それに20 を足した計算をしてxを更新して、表示する
x = 100
x = x + 20
print(x)
```

```python
# チェック演習2-4：
# (1) 変数xに、100を入れ、それに20 を足してyとし、値と型を表示する
x = 100
y = x + 20
print(y)
print(type(y))

# (2) 変数xに、100.0を入れ、それに20 を掛けてyとし、値と型を表示する
x = 100.0
y = x * 20
print(y)
print(type(y))
```

```python
# チェック演習2-5：
# (1) 変数xに、100を入れ、20 で割ってyとし、値と型を表示する。
x = 100
y = x / 20
print(y)
print(type(y))

# (2) 変数xに、100を入れ、それを3で割った答えの整数値をy1とし、さらに、その余りy2を求め、y1, y2の値と型を表示する。
y = x / 3
print(y)
print(type(y))

y1 = x // 3
y2 = x % 3
print(y1, y2)
print(type(y1), type(y2))

# (3) 変数xに、3を入れ、その3乗をyとし、値と型を表示する。
x = 3
y = x **3
print(y)
print(type(y))
```

```python
# チェック演習2-6（累算代入式演算子）：変数x=100 a=20とし、累算代入足し算、累算代入引き算、累算代入掛け算、累算代入割り算 を行い値を表示する。
# 累算代入足し算
x = 100; a = 20
x += a
print(x)

# 累算代入引き算
x = 100; a = 20
x -= a
print(x)

# 累算代入掛け算
x = 100; a = 20
x *= a
print(x)

# 累算代入割り算
x = 100; a = 20
x /= a
print(x)
```
