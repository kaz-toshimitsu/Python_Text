第３章　条件分岐

3.1　比較演算と条件分岐

```python
# コード3.1.1　比較演算(ifのみ）
# 変数ｘに数値10を入れる
x = 10

if x < 5:                    # 比較演算:xは5より小さいか？
    print('xは5より小さい')
```

```python
# コード3.1.2　比較演算 (if) の補足
# 変数ｘに数値10を入れる
x = 10

if x < 5:                    # 比較演算：xは5より小さいか？
    print('xは5より小さい')
    print(x)
```

```python
# 変数ｘに数値10を入れる
x = 10

if x < 5:                    # 比較演算：xは5より小さいか？
    print('xは5より小さい')

print(x)
```

```python
# コード3.1.3　比較演算(if, else）
# 変数ｘに数値10を入れる
x = 10

if x < 5:                    # 比較演算:xは5より小さいか？
    print('xは5より小さい')
else:
    print('xは5以上')
```

```python
# コード3.1.4　比較演算（ifなし)
# 変数ｘに数値1を入れる
x = 1

a = (x < 5)  # 比較演算:xは5より小さいか？
print(a)
```

3.2　論理演算と条件分岐

```python
# コード3.2.1　論理演算子
# 変数ｘに数値10を入れる
x = 10

# xは（5より大きく）かつ（15未満か）
if (x > 5) and (x < 15):
    print('5より大きく、かつ15未満である')
else:
    print('5より大きく、かつ15未満でない')

print(x)
```

3.3　３つ以上の条件分岐

```python
# コード3.3.1　３つ以上の条件分岐
# 変数nに数値1を入れる
x = 1

# nで条件分岐
if x == 0:
    print('x = 0')
elif x == 1:
    print('x = 1')
else:
    print('x > 2')
```
