```python
# コード1.3.1 ノートブックの実行例
# 1 + 2 + ... +10　の計算

# ループカウントNの定義
N = 10

# 計算結果の保存変数[sum]の初期化
sum = 0

# ループ処理
for i in range(1, N + 1):

   # 足し込み計算(sum = sum + i)
    sum += i

   # 結果表示
    print('ループ回数 = ', i, '| 0　～ ', i,  'の足し算 = ', sum)
```
