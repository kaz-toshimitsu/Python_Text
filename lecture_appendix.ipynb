```python
# コード6.1.3 の(表6.2）スライスによる要素範囲の指定
# リストの定義
L = [2, 4, 6, 8, 10]

s = L[ : ]
print(s)

s = L[:3 ]
print('L[:] =', s)

s = L[3: ]
print('L[3: ] =', s)

s = L[-3:]
print('L[-3:] =', s)

s = L[:-1]
print('L[:-1] =', s)

s = L[:-2]
print('L[:-2] =', s)

s = L[::2]
print('L[::2] =',s)

s = L[::-1]
print('L[::-1] =',s)
```

```python
class Grade:
    def __init__(self, name, score):   # オブジェクトが持つ属性と呼ばれる変数を初期設定
        self.name = name    # インスタンス変数
        self.score = score  # インスタンス変数

    def say_name(self):
        print("氏名は", {self.name}, "です。")

# class Gradeでオブジェクト生成
student1 = Grade("情報 花子", 85)
student2 = Grade("情報 一郎", 70)

# class Gradeで属性を確認
print(student1.name)  # => "情報 花子"
print(student2.score) # => 70

# class Gradeで氏名を表示
student1.say_name() # => “氏名は {self.name} です。”
student2.say_name() # => “氏名は {self.name} です。”
```

```python
class Grade:
    def __init__(self, name, score=60):   # オブジェクト初期設定, scoreのデフォルト値を60に設定
        self.name = name
        self.score = score

    def say_name(self):
        print("氏名は", {self.name}, "です。")

student1 = Grade("情報 花子")
student2 = Grade("情報 一郎", 70)

print(student1.score)
print(student2.score)
```
