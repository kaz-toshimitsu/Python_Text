問題７－１

タプルはリストと同じデータ構造で、参照方法も同じです。違う点は、括弧(　)で囲み、要素は変更できないこと

問題７－２

```python
# タプル型の定義
t = ('first', 'second', 'third', 'fourth')
print(t)
# リスト型の性質
print(type(t))
```

問題７－３

```python
# タプルの取り出し
t1 = t[0]
t2 = t[1]
print(t1, t2)
```

問題７－４

```python
#　集合の定義
d = {'first', 'second', 'third', 'fourth', 'first', 'second'}
print(d)
```

問題７－５

```python
# 集合のリストへの変換
l= list(d)
print(l)
```

問題７－６

```python
# 辞書の定義
dic = {'d1':'first', 'd2':'second', 'd3':'third', 'd4': 'fourth'}
print(dic)
# keysメソッド, valuesメソッドの返り値
print(dic.keys())
print(dic.values())
```

問題７－７

```python
# 既存の辞書要素の更新
dic['d2'] = 'second+'
print(dic)
```

問題７－８

```python
# 辞書のvalueに「third」があるか
dic = {'d1':'first', 'd2':'second', 'd3':'third', 'd4': 'fourth'}
value1 = 'third'
print(
    '辞書のvalueに「third」があるか:', value1 in dic.values()
)
```
