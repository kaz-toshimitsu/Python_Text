12.4　データの前処理（データの確認・加工）

```python
# コード12.4.1　データフレームの列の項目名変更
# 差し替え用項目名リスト
cols_E = [
    'gender', 'code',  'prefecture', 'population（>10y）', 'self-learn',
    'self-l(method)', 'English', 'other lang.', 'favorite method', 'commerce'
]

# 加工用の別データフレームの用意
df4 = df3.copy()

# 列名の差し替え
df4.columns = cols_E

# 結果確認
display(df4.head())
```

```python
# コード12.4.2　欠損値の確認（isnullメソッド）
# 各要素ごとにNULLかどうかを判断し、「True」が欠損値
df3.isnull()
```

```python
# コード12.4.3　欠損値の確認（列の欠損値カウント）
# isnullメソッドの結果に対してsumメソッドをかけると、
# 項目単位で何件NULLがあったかがわかる （メソッドチェイン呼び出し)
df3.isnull().sum()
```

```python
# コード12.4.4　欠損値のある行を削除（dropnaメソッド）
# 項目「人口（10歳以上）」の列で欠損値がある行だけを削除する
df5 = df3.copy()
df5 = df5.dropna(subset=['人口（10歳以上）', '学習・自己啓発(方法)', ])
df5.isnull().sum()
df5.isnull()
```

```python
# コード12.4.5　NaNデータを「fillnaメソッド」で、ある値で置き換える
# fillnaの引数(0)は、置き換える値
df51 = df3.fillna(0)

display(df51)
```

*   列項目の移動と削除

```python
# コード12.4.6
# removeで項目「地域コード」を削除する
df6 = df3.copy()
colums1 = list(df6.columns)
colums1.remove('地域コード')
df6 = df6[colums1]
df6.head()
```

```python
# コード12.4.7　列項目の移動と削除
# 元データ項目の「学習・自己啓発」より右の列を左に移動し、「'都道府県'」より右列を全て削除
df7 = df3.copy()
columns1 = list(df7.columns)
print(columns1)
columns2 = columns1[4:] + columns1[:2]
df7 = df7[columns2]
df7.head()
```
