4.2　組込関数の呼び出し

```python
# コード4.2.1　関数①の呼び出し
# 変数ｘ1に数値100.65を入れる
x1 = 100.65

# 変数ｘ1の整数部分をi1に入れる
i1 = int(x1)
print(i1)
```

```python
# コード4.2.2　関数②の呼び出し
# 変数ｘ1に数値100.65を入れる
x1 = 100.65

print(x1)
```

```python
# コード4.2.3　print関数の複数引数とオプション引数
#
i1 = 100
x1 = 3.14

# print関数の複数引数とオプション引数
print(i1, x1)
print(i1, x1, sep = '\n')
```

4.3　よく使う関数と呼び出し

```python
# コード4.3.1 float関数の呼び出し
a = 100
f1 = float(a)
print(f1)
type(f1)
```

```python
# コード4.3.2 str関数の呼び出し
a = 100
s1 = str(a)
print(s1)
type(s1)
```

```python
# コード4.3.3 len関数の呼び出し
s1 = 'abcdefg'
ns1 = len(s1)
print(ns1)
type(ns1)
```

4.4　メソッド

```python
# コード4.4.1 メソッドindexの呼び出し
# 文字列型変数で、text, text1, text2 を指定
text = 'There is the method.'
text1 = 'method'
text2 = 'one'
# 文字列'method'の位置をtextから見つけ、n1に入れる
n1 = text.index(text1)
print(n1)
# 文字列'one'の位置をtextから見つけ、n2に入れる
n2 = text.index(text2)
print(n2)
```

```python
# コード4.4.2 メソッドtitleの呼び出し
# 文字列型変数として、textを指定
text = 'There is the method.'

# 文字列'm1'を全て大文字にする
text3 = text.title()
# 結果確認
print(text3)
```
