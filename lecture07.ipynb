第７章　タプル、辞書、集合

7.1　タプル

```python
# コード7.1.1　タプル型の定義と性質
# タプル型の定義
T = (2, 4, 6, 8, 10)
print(T)
# タプル型の性質
print(type(T))
```

```python
# コード7.1.2　タプルの参照
# タプル型の定義
T = (2, 4, 6, 8, 10)

# タプルの要素の参照
t1 = T[1]
print(t1)

# タプルのスライス参照
print(T[1:3])
```

```python
# コード7.1.3　タプルの要素は変更できない
# タプル型の定義
T = (2, 4, 6, 8, 10)

# タプルの要素の変更しようとしてもエラーがでる
T[1] = 3
```

```python
# コード7.1.4 タプルのアンパック代入
s0, s1, s2, s3, s4 = ('apple', 'orange', 'cherry', 'banana', 'grape')

# アンパック後の各変数の確認
s0, s1, s2, s3, s4
```

7.2　リストにタプル要素入れる（二重構造データとアンパック代入）

```python
# 7.2.1　リストにタプル要素入れる（二重構造データとその書き換え）
# データ定義
d = [('E1', 12), ('E2', 6), ('E3', 4), ('E2', 10)]
# データの確認
print(d)
# 要素の書き換え（最初の組）
d[0] =('F1','G1')
print(d)
```

```python
# 7.2.2　ループ処理でのアンパック代入
# ループ処理で、タプル組要素を取り出す
for key, value in d:
    print('key = ', key, '  value =', value)
```

7.3　集合

```python
# 7.3.1　集合(重複要素は自動的に削除される)
# データ定義
d = {1, 3, 3, 5, 7, 10, 10}

# データの確認
print(d)
```

```python
# 7.3.2　集合要素は参照できない
# データ定義
d = {1, 3, 5, 7, 10}

# データの確認（要素の参照×）
print(d[0])
```

```python
# 7.3.3　集合要素の参照（list関数をつかう）
# データ定義
d = {1, 3, 5, 7, 10}

# データの確認
L = list(d)

print(L)
print(type(L))

print(L[0])
```

```python
# 7.3.4　重複要素のないリストの作成
# 重複要素のあるりストデータの定義
L = [1, 3, 3, 5, 7, 10, 10]

# リスト→（set）→集合：重複要素のない集合データ
d = set(L)
print(d)
# 集合→（list）→リスト：複要素のないりストデータ
L1 = list(d)
print(L1)
print(L1[0])
```

7.4　辞書

```python
# 7.4.1　辞書データ定義
dic = {
    'N5440': 'ノートPC14',
    'N3535': 'ノートPC15',
    'N9340': 'ノートPC16',
    'D5467': 'デスクPCi7'
}

print(dic)
print(type(dic))
```

```python
# 7.4.2　辞書の重複要素の自動削除
# 辞書データ定義
dic = {
    'N5440': 'ノートPC14',
    'N3535': 'ノートPC15',
    'N9340': 'ノートPC16',
    'D5467': 'デスクPCi5',
    'D5467': 'デスクPCi7'
}

print(dic)
```

```python
# 7.4.3　辞書データの参照
print(dic['N5440'])
```

```python
# 7.4.4 辞書要素の追加、更新、削除
# 辞書データ定義
dic = {'N5440': 'ノートPC14', 'N3535': 'ノートPC15'}
print(dic)

# 新しい辞書要素の追加
dic['D5468'] = 'デスクPCi5'
print(dic)

# 既存の辞書要素の更新
dic['N3535'] = 'デスクPCi9'
print(dic)

# 既存の辞書要素の削除
del dic['D5468']
print(dic)
```

```python
# コード7.4.5　辞書内容の確認
# 辞書の定義
dic = {
    'N5440': 'ノートPC14',
    'N3535': 'ノートPC15',
    'N9340': 'ノートPC16',
    'D5467': 'デスクPCi7'
}

# keysメソッド, valuesメソッドの返り値
print(dic.keys())
print(dic.values())

# 辞書のkeyに「D5467」があるか
key1 = 'D5467'
print('辞書のkeyに「D5467」があるか:', key1 in dic.keys())

# 辞書に「N8675」があるか
key2 = 'N8675'
print('辞書に「N8675」があるか:', key2 in dic)

# 辞書のvalueに「デスクPCi7」があるか
value1 = 'デスクPCi7'
print('辞書のvalueに「デスクPCi7」があるか:', value1 in dic.values())
```
