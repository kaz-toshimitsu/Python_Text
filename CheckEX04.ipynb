```python
# チェック演習4-1：「f1=10.5,f2=20.5として、print関数を１度だけ使って、縦にf1,f2の値を表示する。
f1 = 10.5
f2 = 20.5
print(f1, f2, sep='\n')
```

```python
# チェック演習4-2：付録の第4章の組み込み関数を参照して、以下を行う。
# (1) x = [5, 30, -5] として、最大値、最小値、合計求め、表示する。

x = [10, 20, -5]
xmax = max(x)
xmin = min(x)
xsum = sum(x)

print('最大値 = ', xmax)
print('最小値 = ', xmin)
print('合　計 = ', xsum)

# (2) y=14/3を計算し、小数点以下3桁で表示する。
y = round(14/3, ndigits = 3)
print('y=14/3を計算し、小数点以下3桁で表示 = ', y)
```

```python
# チェック演習4-3：
# (1) 英語フレーズ “losing IS NOT an option.”で、先頭の一文字を大文字、他を小文字に変換し、表示する。
s = 'losing IS NOT an option.'
cs = s.upper()
print(s, '=>', cs)

# (2) 英語フレーズ “losing IS NOT an option.”で、”option”が現れる位置のインデックス値を求め、表示する。

i = s.index('option')
print('index number of "option" in "losing IS NOT an option." = ', i)
```
