```python
# 事前の設定

# 必要ライブラリのimport
import numpy as np
# numpy表示形式の設定
np.set_printoptions(
    suppress=True, precision=4, floatmode='fixed'
)
# 日本語化ライブラリ導入
!pip install japanize-matplotlib | tail -n 1
# matplotlib日本語化対応ライブラリのインポート
import japanize_matplotlib
# Matplotlib中のpyplotライブラリのインポート
import matplotlib.pyplot as plt
# グラフのデフォルトフォント指定
plt.rcParams["font.size"] = 12

# サイズ設定
plt.rcParams['figure.figsize'] = (4, 4)
```

```python
# チェック演習10-1：
# コード10.3.4での散布図と同様に、Irisデータを使って、x軸を「花弁（petal）の長さ(cm)」、y軸を「花弁（petal）の幅(cm)」とする散布図を作る。

# グラフ描画用のデータを、ライブラリを用いて取得する
from sklearn.datasets import load_iris
iris = load_iris()
x, y = iris.data, iris.target
columns = iris.feature_names
# columnsの内容
print('column:',columns)

# scatter関数呼び出し　（yの値で色を変える）
plt.scatter(x[:,2], x[:,3], c=y, cmap='rainbow')     #　<=変更

# 方眼表示
plt.grid()

# 軸ラベル表示
plt.xlabel(columns[2])        #　<=変更
plt.ylabel(columns[3])        #　<=変更

# タイトル表示
plt.title('アイリスデータセットによる散布図')

# 描画
plt.show()
```

```python
# コード10.4.2　x-y プロット
# ｘの配列
pi = np.pi    # numpyのπを使用
xRad = np.linspace(0, 2*pi, 61)

# y = exp(x)の計算
ysin = np.sin(xRad) # sin関数
ycos = np.cos(xRad) # cos関数

# plot関数を2度続けて呼び出すと重ね描きになる
# label引数は、凡例表示をする場合に必要
plt.plot(xRad, ysin, label='sin(x)')
plt.plot(xRad, ycos, label='cos(x)')

# 凡例表示
plt.legend()

plt.grid()
plt.xlabel('x (rad)')
plt.ylabel('y')
plt.show()
```

```python
# チェック演習10-3：
# コード10.4.5を参考に、以下の３つの散布図を並べて表示する。
# 散布図１：がく片（sepal）の長さ(cm) VS がく片（sepal）の幅(cm)
# 散布図２：がく片（sepal） の長さ(cm) VS 花弁（petal）の長さ(cm)
# 散布図３：がく片（sepal） の長さ(cm) VS 花弁（petal）の幅(cm)
# 散布図の大きさや配置は自分で調整。

# サイズ指定
plt.figure(figsize=(12, 4))

# グラフ数だけループをする
for i in range(1, 4):

    # i 番目のax変数取得
    ax = plt.subplot(1, 3, i)
    # 散布図表示
    ax.scatter(x[:,0], x[:,i], c=y, cmap='rainbow')
    # タイトル表示
    ax.set_title(columns[0] + ' vs ' + columns[i])
    ax.grid()

# 隣接オブジェクトと重ならないようにする
plt.tight_layout()
# 表示
plt.show()
```
