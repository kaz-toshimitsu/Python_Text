問題４－１

組込関数は、数学で使う関数と同じように「引数」を入れて使用する。
関数①：値を返す関数（結果を変数に代入することができる）
関数②：値以外の方法で結果を返す関数（結果を変数に代入することができない）
がある。
関数①は、type関数、int関数などです。関数②は、print関数などである。形式は「関数名（引数）」で、int(x)のように使う。
