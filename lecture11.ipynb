11.4　データフレームの定義

```python
# コード11.4.1　データフレームのための基礎データの作成
# 2重リスト変数の定義
data_A = [
    ['情報花子', 68, 77, 72.5],
    ['情報一郎', 90, 85, 87.5],
    ['情報二郎', 63, 73, 68.0],
    ['情報三郎', 80, 60, 70.0]
]

# 結果確認
print(data_A)
```

```python
# コード11.4.2 データフレームの定義（関数を使用）
# データフレームの定義

df_A = pd.DataFrame(
    data_A,columns=['氏名', '国語', '英語', '平均']
)

# 型表示
print(type(df_A))
print()

# display関数による整形表示
display(df_A)
```

```python
# コード11.4.3　データフレームの内容確認

# 列インデックスと行インデックス
print('列インデックス', list(df_A.columns)) #columsメソッド
print('行インデックス', list(df_A.index))   #indexメソッド
print()

# データフレームの値
print('データフレームの値')
print(df_A.values)   #valuesメソッド
```

Series変数

```python
# コード11.4.4 リスト変数からSeries変数を作る
s1 = pd.Series( [68, 90, 63, 80], name='国語')

print(type(s1))
print(s1)
```

```python
# コード11.4.5　データフレームの特定列からSeries変数を抽出
s2 = df_A['国語']

print(type(s2))
print(s2)
```

```python
# コード11.4.6　データフレームの特定行からSeries変数を抽出
s3 = df_A.loc[1]

print(type(s3))
print(s3)
```

11.5　データフレームの値の抽出

```python
# コード11.5.1　各項目のデータ型確認
print(df_A.dtypes)
```

```python
# コード11.5.2 列の一部の抽出
# 抽出したい項目名のリストを準備
column1 = ['国語', '英語']

#  <変数名>　[cols]の形で列抽出
df_B = df_A[column1]

# 結果確認
display(df_B)
```

```python
# コード11.5.3　行の最初の部分の抽出（headメソッド）
# head(n)メソッドで先頭n行の抽出（指定しない場合は、デフォルト値=5）
display(df_A.head(2))
```

```python
# コード11.5.4　行の最後の部分の抽出（tailメソッド）
# tail(n)メソッドで最後のn行の抽出（指定しない場合は、デフォルト値=5）
display(df_A.tail(2))
```

```python
# コード11.5.5 行の一部の抽出（条件指定）
# index1: 「平均が70以上]の判断
index1 = df_A['平均'] >= 70
print(index1)
print()
# index1で行を絞り込む
df_C = df_A[index1]
# 表示
display(df_C)
```

```python
# コード11.5.6　行の一部の抽出（短縮表記）
# 「平均が70以上]の判断(短縮して1行で行う)
df_D = df_A[df_A['平均'] >= 70]

# 表示
display(df_D)
```

11.6　データフレームの列・行の追加や削除

```python
# コード11.6.1　列の削除
# 項目「氏名」の列を削除する
# データをコピーしてオリジナルに影響がないようにする
df_A0 = df_A.copy()

# 項目「氏名」を削除し、結果をdf_Eに代入
df_E = df_A0.drop('氏名', axis=1)

# 結果を確認
display(df_E)
```

```python
# コード11.6.2　列の追加（最右欄）
# 合計点の計算
s1 = df_A['国語'] + df_A['英語']

# 結果確認
print(s1)
```

```python
# 合計点列の追加

# データコピー
df_F = df_A.copy()

# 列追加
df_F['合計点'] = s1

# 結果確認
display(df_F)
```

```python
# コード11.6.2　列の追加（insert関数により任意の列位置）
# データコピー
df_G = df_A.copy()

# 列追加
df_G.insert(1, '合計点', s1)

# 結果確認
display(df_G)
```

```python
# コード11.6.3　既存データフレームに新しい列を追加（concat関数）
# 新しいデータ「学年」
s2 = pd.Series(np.array([2, 2, 1, 1]), name='学年')
print(s2)
```

```python
# 既存データフレームに新しい列を追加
df_H = pd.concat([df_G, s2], axis=1)

# 結果確認
display(df_H)
```

```python
# コード11.6.4　新しい行を最後に追加する（concat関数）
# 追加する学生のデータ
new_row = {'氏名': '追加学生', '国語': 30, '英語': 70, '平均': 50.0}

# 既存データフレーム「df_A」に追加
df_I = pd.concat([df_A, pd.DataFrame([new_row])], ignore_index=True)
# 結果確認
display(df_I)
```

```python
# コード11.6.5　列項目をindexにする
df_A.set_index('氏名', inplace = True) #
display(df_A)
```

```python
# コード11.6.6　連番のインデックスに戻す
df_A.reset_index('氏名', inplace = True)
display(df_A)
```
