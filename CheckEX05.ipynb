```python
# チェック演習5-1：
# (1) 以下に示す、非構造化プログラムを構造化プログラムにする。
# 非構造化プログラム

a = 10; b = 20; c = 30; d = 40

x1 = a * b
x2 = c * d
y = x1 * x2

print(y)
```

```python
# 構造化プログラム
# --- 定義関数
def func(a, b):
    x = a * b
    return x

# --- main program ---
a = 10; b = 20; c = 30; d = 40

x1 = func(a, b)
x2 = func(c, d)
y  = func(x1, x2)
print(y)
```

```python
# (2) 上の(1)で作成した掛け算の関数を、足し算に変える。
# 構造化プログラム
# --- 定義関数 x = a * b を　x = a + b にすることで、一括変更できる。
def func(a, b):
    x = a + b
    return x

# --- main program ---
a = 10; b = 20; c = 30; d = 40

x1 = func(a, b)
x2 = func(c, d)
y  = func(x1, x2)
print(y)
```

```python
# チェック演習5-2：関数プログラムとして、2変数の掛け算、aのb乗、bのa乗の３つの戻り値を定義し、戻り値の合計を求めるプログラムを作る。
# --- 定義関数
def func(a, b):
    x1 = a * b
    x2 = a ** b
    x3 = b ** a
    return x1, x2, x3

# --- main program ---
y1 = 10; y2 = 5
y  = sum(func(y1, y2))
print(y)
```
