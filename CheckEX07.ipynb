```python
# チェック演習7-1：
# (1) 'apple', 'orange', 'cherry', 'banana', 'grape'が要素のタプルT71を作って、表示する。

T71 = ('apple', 'orange', 'cherry', 'banana', 'grape')
print(T71)
```

```python
# (2) インデックス0番目の要素を変更しようとして、エラーになることを確認する。

T71[0] = 'Apple'
```

```python
# チェック演習7-2：チェック演習7-2で作ったタプルT71のアンパック代入を行い、表示する。
T720, T721, T722, T723, T724 = T71
print('T71をアンパック代入し0番目の要素を表示 = ', T720)
print('T71をアンパック代入し1番目の要素を表示 = ', T721)
print('T71をアンパック代入し2番目の要素を表示 = ', T722)
print('T71をアンパック代入し3番目の要素を表示 = ', T723)
print('T71をアンパック代入し4番目の要素を表示 = ', T724)
```

```python
# チェック演習7-3：
# (1) 下記の２重構造のタプルで構成するリストL73を作り、表示する。
# (0, 'apple'), (1,'orange'), (2, 'cherry')
L73 = [(0, 'apple'), (1,'orange'), (2, 'cherry')]
print(L73)

# (2) 0番目の２重構造のタプル要素を表示し、(0, 'Apple')に変更する。
L73[0] = (0, 'Apple')
print(L73)

# (3) ループ処理でL73のタプル要素を取り出し、表示する。
for key, value in L73:
    print('key = ', key, ' value =', value)
```

```python
# チェック演習7-4：下記の要素で集合d74を作り、重複する要素がない結果を表示する。
# 'a', 'b', 'c', 'd', 'b', 'd', 'd'

d74 = {'a', 'b', 'c', 'd', 'b', 'd', 'd'}
print(d74)
```

```python
# チェック演習7-5：チェック演習7-4で作った集合をリストに変換する。

# 集合→（list）→リスト
L75 = list(d74)
print(L75)
```

```python
# チェック演習7-6：次で示す商品管理リストの辞書データdic76を作り、表示する。
# F1:クルミ, F2:リンゴ, F3:ミカン, F4:イチゴ

dic76 = {'F1':'クルミ', 'F2':'リンゴ', 'F3':'ミカン', 'F4':'イチゴ'}
print(dic76)

print(dic76['F2'])
```

```python
# チェック演習7-7：
# (1) チェック演習7-6で作った辞書に新しい辞書要素(F5:カキ） を追加する。
dic76['F5'] = 'カキ'
print(dic76)

# (2) 要素F2をマロンに更新する。
dic76['F2'] = 'マロン'
print(dic76)

# (3) 要素F4を削除する。
del dic76['F4']
print(dic76)
```

```python
# チェック演習7-8：
# (1) チェック演習7-7の最後の辞書データをdic78として、keysとvaluesを確認する。
dic78 = dic76
print('辞書データ: ', dic78)
print(dic78.keys())
print(dic78.values())

# 辞書dic78のkeyに「F4」があるか
key1 = 'F4'
print('辞書のkeyに「F4」があるか:', key1 in dic78.keys())

# 辞書のvalueに「マロン」があるか
value1 = 'マロン'
print('辞書に「マロン」があるか:', value1 in dic78.values())
```
